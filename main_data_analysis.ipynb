## The Dickey-Fuller Test
----
In this section we handle checking to see if the data is stationary for each of our states and each of our parameters. This is an important test to determine the degree of differencing that we need in our later SARIMAX modeling. Let's start by getting the required modules from the `statsmodels` library.

```python
from statsmodels.tsa.stattools import adfuller
```

Now, we can write a function to compute the needed results from the Dickey Fuller test on each state, and wrap these to an easily readable or mutable dataframe object.

```python
def compute_state_adfuller(state_name):
    subset = main_df.loc[main_df.state_name == state_name]
    p_row_to_add = {}
    nlags_row_to_add = {}
    test_stat_row_to_add = {}
    for param in subset.parameter_code.unique():
        param_aqi = subset.loc[subset.parameter_code == param].set_index('date_local').aqi
        param_ad_fuller_results = adfuller(param_aqi, autolag='AIC')
        test_stat_row_to_add[param] = param_ad_fuller_results[0]
        p_row_to_add[param] = param_ad_fuller_results[1]
        nlags_row_to_add[param] = param_ad_fuller_results[2]
        
        
    results_df = pd.DataFrame(
        [test_stat_row_to_add, p_row_to_add, nlags_row_to_add],
        index=['test_stat', 'p_value', 'n_lags'])
    results_df.columns.name = 'parameter_code'
    results_df = results_df.T
    
    results_df['state_name'] = state_name
    results_df = results_df.set_index(['state_name', results_df.index])
    return results_df
```

We can now use our function to try computing the results for all 50 states, saving the results in a larger dataframe.

```python
ad_fuller_results = pd.DataFrame()
for state_name in main_df.state_name.unique():
    ad_fuller_results = pd.concat([ad_fuller_results, compute_state_adfuller(state_name)])
```

We may want to refer to this data at a later point without having to recompute the Dickey-Fuller test on each state, so we will save it as a CSV in the appropriate [`results_data`](./results_data/) folder.

## SARIMAX Fitting
----
In this section, we aim to fit a SARIMAX model to our states and parameters, so that we can make some predictions for the year 2019. Let's start by writing a function that will fit SARIMAX models for a given state by name. We should also import some metrics that we will use to evaluate the model.

```python
from sklearn.metrics import mean_squared_error
```

```python
def compute_aqi_pred_mse(aqi_series, fitted_model, seasonal_freq):
    model_pred = fitted_model.predict()
    aqi = aqi_series.iloc[seasonal_freq:]
    pred = model_pred.iloc[seasonal_freq:]
    mse = mean_squared_error(aqi, pred)
    return mse

def fit_state_monthly_sarimax(state_name, order=(1,0,2), seasonal_order=(0,1,1,12)):
    subset = main_df.loc[main_df.state_name == state_name]
    aic_row = {}
    bic_row = {}
    rmse_row = {}
    fitted_models = {}
    for param in subset.parameter_code.unique():
        param_subset = subset.loc[subset.parameter_code == param]
        param_subset_monthly_resample = param_subset.set_index('date_local').resample('MS').mean()
        mod = sm.tsa.statespace.SARIMAX(
            param_subset_monthly_resample.aqi,
            trend='n',
            order=order,
            seasonal_order=seasonal_order)
        results = mod.fit()
        aic_row[param] = results.aic
        bic_row[param] = results.bic
        try:
            mse = compute_aqi_pred_mse(param_subset_monthly_resample.aqi, results, seasonal_order[3])
            rmse_row[param] = np.sqrt(mse)
        except ValueError:
            mse = np.nan
            rmse_row[param] = mse
        fitted_models[(state_name, param)] = results
    
    results_df = pd.DataFrame([aic_row, bic_row, rmse_row], index=['aic', 'bic', 'rmse'])
    results_df.columns.name = 'parameter_code'
    results_df = results_df.T
    results_df['state_name'] = state_name
    results_df.set_index(['state_name', results_df.index], inplace=True)
    return results_df, fitted_models
```

Now that we have this function which can both get the important results as a Pandas dataframe as well as store the fitted models in a dictionary object indexed by tuples representing (state, parameter) combinations, we can loop over all the possible states to find all relevant metrrics for monthly models, as well as store the full set of fitted models in a Python dictionary for easy future prediction.

```python
all_state_monthly_sarimax_results = pd.DataFrame()
all_state_monthly_models = {}

for state_name in main_df.state_name.unique():
    new_results, new_models = fit_state_monthly_sarimax(state_name)
    all_state_monthly_sarimax_results = pd.concat(
        [all_state_monthly_sarimax_results, new_results])
    all_state_monthly_models.update(new_models)
```

## SARIMAX Forecasting
----
Now the we have fitted some models to our observed data, we can try to use these data to forecast into 2019 and beyond. Let's start by writing a function that computes predictions and forecasts for a given number of months within a given state. This will make it much easier to plot or perform validation using our predicted and forecasted data.

```python
from dateutil.relativedelta import relativedelta

def make_state_aqi_forecasts(state_name, n_periods, resampling_freq='MS'):
    results_df = pd.DataFrame()
    state_subset = main_df.loc[main_df.state_name == state_name]
    for param in state_subset.parameter_code.unique():
        param_subset = state_subset.loc[state_subset.parameter_code == param]
        param_resample = param_subset.set_index('date_local').resample(resampling_freq).mean()
        fitted_model = all_state_monthly_models[(state_name, param)]
        model_pred = fitted_model.predict()
        df_to_append = pd.DataFrame({'aqi': param_resample.aqi, 'pred_aqi': model_pred}, index=param_resample.index)
        start_index = param_resample.index[-1]+relativedelta(months=1)
        new_dates_to_add = [start_index + relativedelta(months=n) for n in range(n_periods)]
        future = pd.DataFrame(index=new_dates_to_add, columns=df_to_append.columns)
        df_to_append = pd.concat([df_to_append, future])
        df_to_append['fore_aqi'] = fitted_model.predict(start=start_index, end=new_dates_to_add[-1])
        new_col_tup = [(param, c) for c in df_to_append.columns]
        new_cols = pd.MultiIndex.from_tuples(new_col_tup)
        df_to_append.columns = new_cols
        
        results_df = pd.concat([results_df, df_to_append], axis=1)
    return results_df
```

Now, we can use this function to produce predictions and forecasts for all 50 states for the next three years. These results will be saved in CSV format in the [`results_data`](./results_data/) folder.

```python
n_months = 36

for state in main_df.state_name.unique():
    state_forecasts = make_state_aqi_forecasts(state, n_months)
    state_forecasts.to_csv(f'results_data/{state}_{n_months}_mo_forecast.csv')
```

Next, we may want to plot these predictions and forecasts against each state's observed values. Let's write a function that will help us to do this easily.

```python
def plot_predictions_and_forecasts(true_results, predictions, forecast, figsize=(18, 12)):
    fig, ax = plt.subplots(figsize=figsize)
    sns.lineplot(data=true_results, ax=ax)
    sns.lineplot(data=predictions, ax=ax)
    sns.lineplot(data=forecast, ax=ax)
    ax.set_xlabel('Date')
    ax.set_ylabel('Air Quality Index')
    ax.legend(['True Values', 'Model Predictions', 'Forecasted Values'])
    return fig, ax
```

Now we will write a wrapper function for the above to handle the plotting for each parameter within each state. These results plots will be saved to the [`plots/forecast`](./plots/forecast/) folder.

```python
def plot_state_monthly_predictions_and_forecasts(state_name, forecast_months, seasonal_freq=12, figsize=(18, 12)):
    forecast_df = make_state_aqi_forecasts(state_name, forecast_months)
    param_list = list(forecast_df.columns.get_level_values(0).unique())
    for param in param_list:
        param_subset = forecast_df[param]
        param_subset = param_subset.iloc[seasonal_freq:,:]
        aqi = param_subset.aqi.dropna()
        pred_aqi = param_subset.pred_aqi.dropna()
        fore_aqi = param_subset.fore_aqi.dropna()
        p_fig, p_ax = plot_predictions_and_forecasts(aqi, pred_aqi, fore_aqi, figsize=figsize)
        legend_entry = aqi_gasses.loc[aqi_gasses.code == param].legend_entry.values[0]
        p_ax.set_title(f'True Results vs Model Predicted and Forecasted Values - {legend_entry} in {state_name}')
        p_fig.savefig(f'plots/forecast/{state_name}_{param}_forecasted.png')
        plt.close(p_fig)
```

Finally, let's generate our plots using the for loop below.

```python
n_months = 36
for state in main_df.state_name.unique():
    plot_state_monthly_predictions_and_forecasts(state, n_months, figsize=(18, 10))
```

After we have produced the forecasts for the next three years, we may want to compare our models using a basic form of validation. In particular, there is some data that is available through the EPA website on measurements of air pollutants in 2019. We will import this data and perform the same cleaning procedures applied to our training data, so let's copy some needed functionality from the [data_cleaning](data_cleaning.ipynb) notebook.

```python
data_folder = 'raw_data'

cols_to_drop = [
    'POC',
    'Latitude',
    'Longitude',
    'Datum',
    'Pollutant Standard',
    'Event Type',
    'Observation Count',
    'Observation Percent',
    '1st Max Value',
    '1st Max Hour',
    'Method Code',
    'Method Name',
    'Address',
    'CBSA Name',
    'Date of Last Change',
    'Local Site Name']

def import_aggregate_data(param_code, year):
    path_stub = f'/daily_{param_code}_{year}.zip'
    path = data_folder+path_stub
    df = pd.read_csv(path)
    df.drop(columns=cols_to_drop, inplace=True)
    df.columns = df.columns.str.lower().str.replace(' ', '_')
    df.date_local = pd.to_datetime(df.date_local)
    return df
```

Now, let's write a function that will handle producing a dataframe of metrics for a given state's models. This will compute the mean-squared-error and root-mean-squared-error for the forecasted values and available 2019 observed data.

```python
def validate_state_aqi_forecast_2019(state_name):
    forecast_df = make_state_aqi_forecasts(state_name, 12)
    param_list = list(forecast_df.columns.get_level_values(0).unique())
    results_df = pd.DataFrame()
    for param in param_list:
        param_subset = forecast_df[param]
        param_2019 = import_aggregate_data(param, 2019)
        param_2019 = param_2019.dropna(subset=['aqi'])
        daily_means = param_2019.groupby(['state_name', 'parameter_code', 'date_local']).mean()
        daily_means = daily_means.drop(columns=['county_code', 'site_num', 'state_code', 'arithmetic_mean'])
        state_subset_2019 = daily_means.loc[state_name, :]
        param_subset_2019 = state_subset_2019.loc[param,:]
        param_subset_2019_resample = param_subset_2019.resample('MS').mean()
        comp_df = pd.concat([param_subset, param_subset_2019_resample], axis=1, join='inner').dropna(axis=1)
        mse = mean_squared_error(comp_df.aqi, comp_df.fore_aqi)
        results_df.loc[param, 'mse'] = mse
    results_df.index = results_df.index.rename('parameter_code')
    results_df['rmse'] = results_df.mse.map(np.sqrt)
    return results_df
```

Let's now use this function to create a dataframe with results for all states that have available results. Note that we need to except KeyErrors as there are a few states that seem to be lacking in some data, creating issues with accessing data in the 2019 results.

```python
validation_results = pd.DataFrame()
for state in main_df.state_name.unique():
    try:
        new_data = validate_state_aqi_forecast_2019(state)
        new_data['state_name'] = state
        new_data = new_data.set_index(['state_name', new_data.index])
        validation_results = pd.concat([validation_results, new_data])
    except KeyError:
        pass

validation_results    
```
